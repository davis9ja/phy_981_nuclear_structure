### HW13 question 3

What are the allowed J values for 19O in the d5/2 model space? What are
the spectroscopic factors from the 20O ground state to each of these states
in 19O?

Allowed values for 19O in d5/2 model space: J = 3/2, 5/2, 9/2

20O has 4 neutrons in the sd shell, and a ground state J = 0+. There are three possible transfers from 20O that can occur in d5/2 model space, namely the allowed values of J for 19O. The possible final J values come from the triangle condition of 0+,(3/2,5/2,9/2). The spectroscopic factor comes from the number of neutrons that can be transfered from 20O to 19O. d5/2 restricts considerations to final states of J=5/2. The only 19O J state that can couple with the 20O 0+ state to reach J=5/2 is, of course, the 5/2 in 19O.

Therefore, the spectroscopic factors go like this:

|J$_i$|J$_f$|S|
|---|---|-|
|0|3/2|0|
|0|5/2|4|
|0|9/2|0|
